```python
product_categories = ("Electronics", "Stationery", "Clothing")

```

```python
products = ["Laptop", "Notebook", "T-shirt"]

```

```python
new_product = input("Enter a new product to add: ")
products.append(new_product)

```

```python
print(products)
```

```python
# Step 4: Remove a product by index


remove_index = int(input(f"Enter the index (0 to {len(products)-1}) of the product to remove: "))
if 0 <= remove_index < len(products):
    del products[remove_index]
else:
    print("Invalid index. No product removed.")

```

```python
# Step 5: Display results
print("\n Final Product Categories (Fixed):")
for category in product_categories:
    print("-", category)

```

```python
print("\n Final Product List (Updated):")
for index, product in enumerate(products, start=1):
    print(f"{index}. {product}")

```
