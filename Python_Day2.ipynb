```python
feedback = input("Please enter your feedback: ")

```

```python
word_count = len(feedback.split())
print(f"Total number of words: {word_count}")

```

```python
data_count = feedback.lower().count("data")
print(f"The word 'data' appears {data_count} time(s).")

```

```python

first_five = feedback[:5]
last_five = feedback[-5:]
print(f"First 5 characters: '{first_five}'")
print(f"Last 5 characters: '{last_five}'")

```

```python
reversed_feedback = feedback[::-1]
print(f"Reversed feedback: '{reversed_feedback}'")

```
